```python
import numba 
from numba import cuda, jit, njit
import numpy as np
from functools import reduce
import datetime as dt
import matplotlib.pyplot as plt

def now():
    return dt.datetime.now()

with open('offline_1000000.csv', 'r', encoding='utf-8') as arq:
    array = np.array([int(i) for i in arq.read().split('\n') if i])
ref = max(array)
array.size, array.dtype, ref
```

```python
@njit(parallel=True)
def maximo(array):
    return array.max()
times = []
for i in range(1000):
    t0 = dt.datetime.now()
    m = maximo(array)
    times.append((dt.datetime.now()-t0).total_seconds())
times = sorted(times)
media = np.average(times)
print(maximo(array)==ref, media)
plt.figure(figsize=(30,3))
plt.plot(times, label='Tempos')
plt.plot(len(times) *[media], label='MÃ©dia')
plt.title('MÃ©dia:' + str(media))
plt.legend()
plt.yscale('log')
plt.show()
```

```python
@cuda.jit()
def maximo0(array, max_gpu):
    local_max = max_gpu[1]
    for i in range(cuda.grid(1), array.shape[0], cuda.gridsize(1)):
        local_max = max(array[i], local_max)
    cuda.atomic.max(max_gpu, 0, local_max)

def maximo(array):
    array_gpu = cuda.to_device(array)
    max_gpu = cuda.to_device(np.zeros(1, dtype=np.float64))
    maximo0[1000, 1000](array_gpu, max_gpu)
    return max_gpu.copy_to_host()[0]
times = []
for i in range(1000):
    t0 = dt.datetime.now()
    m = maximo(array)
    times.append((dt.datetime.now()-t0).total_seconds())
times = sorted(times)
media = np.average(times)
print(maximo(array)==ref, media)
plt.figure(figsize=(30,3))
plt.plot(times, label='Tempos')
plt.plot(len(times) *[media], label='MÃ©dia')
plt.title('MÃ©dia:' + str(media))
plt.legend()
plt.yscale('log')
plt.show()
```
